```python
#Criar novo DataFrame e gerar média

import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['Kaio', 'Kivia', 'Cleide', 'Carlos', 'Kaue'],
    'salario': [np.nan, 4000, np.nan, 6000, 10000]
})
#Calcula a média e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#Calcula mediana e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Kaio', 'Kivia', 'Cleide', 'Carlos', 'Kaue'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Uberlandia']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

# Aula 3 - Graficos e Histórias com Dados


```python
df_limpo.head()
```

```python
df_limpo['experience_level'].value_counts().plot(kind='bar', title='Distribuição de cargo')
```

```python
#Importação de biblioteca gráfica

import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 2))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
plt.title('Salario médio por nivel de Senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salario médio anual em dólares')
plt.show()
```

```python
df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=True).index
ordem
```

```python
plt.figure(figsize=(10, 2))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd', order = ordem)
plt.title('Salario médio por nivel de Senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salario médio anual em dólares')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df_limpo['salary_in_usd'], bins=50, kde=True)
plt.title('Distribuição dos salários anuais')
plt.xlabel('Salario em USD')
plt.ylabel('Frequencia de cada salário')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=df_limpo['salary_in_usd'])
plt.title('Boxplot Salario')
plt.xlabel('Salario em USD')
```
